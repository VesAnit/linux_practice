```python
import time

def measure_time(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f'Время выполнения функции {func.__name__}: {end - start: .6f} секунд.')
        return result
    return wrapper

@measure_time
def func_wait():
    print('В процессе')
    time.sleep(3)
    print('Готово!')

func_wait()
```

```python
def check_access(role_required):
    def decor(func):
        def wrapper(user_role, *args, **kwargs):
            if user_role == role_required:
                return func(*args, **kwargs)
            else:
                print(f'Доступ к функции {func.__name__} запрещен для роли {user_role}')
                return None
        return wrapper
    return decor


@check_access("admin")
def delete_user(user_id):
    return f'Пользователь с ID {user_id} удален.'


print(delete_user('admin', 123))
print(delete_user('Ivan', 567))
```
